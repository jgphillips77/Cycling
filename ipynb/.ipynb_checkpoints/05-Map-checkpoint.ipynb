## Where did we ride?

```python
# Find range of Latitude & Longitude values.
lap3['latitude'].max(), lap3['latitude'].min(), lap3['latitude'].mean()
```

```python
lap3['longitude'].max(), lap3['longitude'].min(), lap3['longitude'].mean()
```

```python
# Plot Lap3
fig = plt.figure(figsize=(16, 16))
cell = Basemap(projection='lcc', resolution='f', 
            lat_0=36.9, lon_0=136.7,
            width=1.0E6, height=1.0E6)
cell.shadedrelief(scale=0.5)
cell.drawcoastlines(color='k')
cell.etopo()
cell.bluemarble()

#Plot Tokyo.
x_nrt, y_nrt = cell(139.6917, 35.6895)
plt.plot(x_nrt, y_nrt, 'ok', marker="*", markersize=9, color='red')
plt.text(x_nrt, y_nrt, 'Tokyo', fontsize=15, color='red')

#Plot Toyama, closest city.
x_toy, y_toy = cell(137.2137, 36.6960)
plt.plot(x_toy, y_toy, 'ok', markersize=4, color='red')
plt.text(x_toy, y_toy, 'Toyama', fontsize=13, color='red')


lat = lap3['latitude'].values
lon = lap3['longitude'].values
cell.scatter(lon, lat, latlon=True,c='r')

plt.show()
```

The race course was in the Ishikawa Prefecture of Japan.
