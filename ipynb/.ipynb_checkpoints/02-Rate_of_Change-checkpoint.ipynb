```python
# Import pickled clean data:
lap1 = pd.read_pickle('data/lap1.p')
lap2 = pd.read_pickle('data/lap2.p')
lap3 = pd.read_pickle('data/lap3.p')
```

## Look at the difference in Foot AR Q1

```python
# Start with making a new column for Percent Difference
lap1['p_diff'] = 100*(lap1['pedaling_delay_log_left'] - lap1['pedaling_delay_log_right'])/lap1['pedaling_delay_log_left']
lap2['p_diff'] = 100*(lap2['pedaling_delay_log_left'] - lap2['pedaling_delay_log_right'])/lap2['pedaling_delay_log_left']
lap3['p_diff'] = 100*(lap3['pedaling_delay_log_left'] - lap3['pedaling_delay_log_right'])/lap3['pedaling_delay_log_left']
```

```python
lap1['p_diff'].plot()
plt.show()
```

```python
lap2['p_diff'].plot()
plt.show()
```

```python
lap3['p_diff'].plot()
plt.show()
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3)
sns.boxplot(y=lap1['p_diff'], ax=ax1)
sns.boxplot(y=lap2['p_diff'], ax=ax2)
sns.boxplot(y=lap3['p_diff'], ax=ax3)
plt.tight_layout()

# Notice the greater differences in Lap 3.
```

## Rate of Change

```python
# Compute correlation of some items
correlation1 = lap1['power'].corr(lap1['torso_d_rot_z_center'])
print("Correlation of levels: ", correlation1)

# Compute correlation of percent changes
changes1 = lap1.pct_change()
correlation12 = changes1['power'].corr(changes1['torso_d_rot_z_center'])
print("Correlation of changes: ", correlation12)
```

```python
changes1['torso_d_rot_z_center'].plot()
plt.show()
```
