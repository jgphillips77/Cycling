```python
# Import pickled clean data:
lap1 = pd.read_pickle('data/lap1.p')
lap2 = pd.read_pickle('data/lap2.p')
lap3 = pd.read_pickle('data/lap3.p')
```

```python
# Concatenate all three laps for better modeling

data = pd.concat([lap1, lap2, lap3], axis=0)
```

```python
# Standard plot to inspect the join.
plt.figure(figsize=(8, 8))
data.power.plot()
data.altitude.plot()
data.cadence.plot()
data.speed.plot()
plt.show()
```

```python
# Let's do a model focused on power.
X = data.drop(['power'], axis=1) 

# Make a y based upon Power
y = data['power']
```

```python
# Standard split:
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state = 42)
```

## Random Forest Regressor

```python
# Use the best hyperparameters from before to fit a random forest model
rfr = RandomForestRegressor(n_estimators=200, max_depth=5, max_features=12, random_state=42)
rfr_model = rfr.fit(X_train, y_train)

print(rfr.score(X_train, y_train))
print(rfr.score(X_test, y_test))
```

```python
# Get feature importances from our random forest model
importances = rfr.feature_importances_
feature_names = X.columns

# Get the index of importances from greatest importance to least
sorted_index = np.argsort(importances)[::-1]
x = range(len(importances))

# Create labels 
labels = np.array(feature_names)[sorted_index]
plt.figure(figsize=(8, 8))
plt.barh(x, importances[sorted_index], tick_label=labels)
plt.xticks(rotation=90)
plt.show()
```

## Random Forest Permutation Importance

```python
import eli5
from eli5.sklearn import PermutationImportance
```

```python
# Create PI instance 
perm = PermutationImportance(rfr_model, random_state=42).fit(X_test, y_test)

# Visualize Results
eli5.show_weights(perm, feature_names = X_test.columns.tolist())
```

## What about Boosting?

```python
# Create Gradient Boost Model
gbr = GradientBoostingRegressor(max_features=12, learning_rate=0.02, n_estimators=200,
                                subsample=0.6, random_state=42)
gbr_model = gbr.fit(X_train, y_train)

print(gbr.score(X_train, y_train))
print(gbr.score(X_test, y_test))
```

```python
# Extract feature importances from the fitted gradient boosting model
feature_importances = gbr.feature_importances_
features = X.columns

# Get the indices of the largest to smallest feature importances
sorted_index = np.argsort(feature_importances)[::-1]
x = range(features.shape[0])

# Create tick labels 
labels = np.array(feature_names)[sorted_index]
plt.figure(figsize=(8, 8))
plt.barh(x, feature_importances[sorted_index], tick_label=labels)
plt.xticks(rotation=90)
plt.show()
```

## Gradient Boost Permutation Importance

```python
# Create PI instance 
perm2 = PermutationImportance(gbr_model, random_state=42).fit(X_test, y_test)

# Visualize results
eli5.show_weights(perm2, feature_names = X_test.columns.tolist())
```
