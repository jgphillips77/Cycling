```python
# Import pickled clean data:
lap1 = pd.read_pickle('data/lap1.p')
lap2 = pd.read_pickle('data/lap2.p')
lap3 = pd.read_pickle('data/lap3.p')
```

## Heatmaps to look for some correlations.

```python
# Lap1 Heatmap.
corr_lap1 = lap1.corr(method='pearson')
plt.figure(figsize=(16,16))
sns.heatmap(corr_lap1, annot=True, cmap='icefire')
```

```python
# Lap2 Heatmap.
corr_lap2 = lap2.corr(method='pearson')
plt.figure(figsize=(16,16))
sns.heatmap(corr_lap2, annot=True)
```

```python
# Lap3 Heatmap.
corr_lap3 = lap3.corr(method='pearson')
plt.figure(figsize=(15,15))
sns.heatmap(corr_lap3, annot=True, cmap='hot_r')
```

## Power

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3)
sns.boxplot(y=lap1['power'], ax=ax1)
sns.boxplot(y=lap2['power'], ax=ax2)
sns.boxplot(y=lap3['power'], ax=ax3)
plt.tight_layout()
```

```python
# Standard Strava-ish plots
lap1.power.plot()
lap1.altitude.plot()
lap1.cadence.plot()
lap1.speed.plot()
plt.axhspan(296, 354, alpha=0.3, color='grey') #Shaded area for 25-75 percentile range of Power
plt.show()
```

```python
lap2.power.plot()
lap2.altitude.plot()
lap2.cadence.plot()
lap2.speed.plot()
plt.axhspan(294, 359, alpha=0.3, color='grey') #Shaded area for 25-75 percentile range of Power
plt.show()
```

```python
lap3.power.plot()
lap3.altitude.plot()
lap3.cadence.plot()
lap3.speed.plot()
plt.axhspan(304, 374, alpha=0.3, color='grey') #Shaded area for 25-75 percentile range of Power
plt.show()
```

## Waist Angle per lap

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3)
sns.violinplot(y=lap1['waist_angle_log_center'], ax=ax1)
sns.violinplot(y=lap2['waist_angle_log_center'], ax=ax2)
sns.violinplot(y=lap3['waist_angle_log_center'], ax=ax3)
plt.tight_layout()

# Definite increase on Lap 3!
```
