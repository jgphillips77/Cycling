## Data Cleaning

```python
# How many null values do we have?
lap1.isnull().sum().nlargest(15).plot(kind="barh")
```

```python
lap2.isnull().sum().nlargest(15).plot(kind="barh")
```

```python
lap3.isnull().sum().nlargest(15).plot(kind="barh")
```

```python
# Today for simplicity sake, I will drop the columns full of null values:
# But I really want to know why this column is full of NaNs.  Would love to see those metrics!
lap1.drop(['l_r_balance', 'heartrate', 'leg_range_log_left', 'leg_range_log_right'], axis=1, inplace=True)
lap2.drop(['l_r_balance', 'heartrate', 'leg_range_log_left', 'leg_range_log_right'], axis=1, inplace=True)
lap3.drop(['l_r_balance', 'heartrate', 'leg_range_log_left', 'leg_range_log_right'], axis=1, inplace=True)
```

```python
# Next I will fill the remaining null values.
lap1.fillna(method='bfill', inplace=True)
lap2.fillna(method='bfill', inplace=True)
lap3.fillna(method='bfill', inplace=True)
```

```python
# How is our cleaning working?
lap1.isnull().sum()
```

```python
lap2.isnull().sum()
```

```python
lap3.isnull().sum()
```

```python
# Use Elapsed Time as the index
lap1.set_index('elapsed_time', inplace=True)
lap2.set_index('elapsed_time', inplace=True)
lap3.set_index('elapsed_time', inplace=True)
```

## Scatter plots to look at distribution:
* For simplicity today, I'm intersted in Power, and how the other metrics potentially contribute to better Power.

```python
# Numeric Columns that I want to keep for plots:
lap1_num = lap1.select_dtypes(exclude=['object', 'category'])  
```

```python
# Scatter Subplot of Numeric Columns
fig = plt.figure(figsize=(16,16))

#Make for loop that will plot Numeric Columns vs Power

for i, col in enumerate(lap1_num):
    fig.add_subplot(10,4, 1+i)
    plt.scatter(lap1_num[col], lap1['power'])
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('Power')    

plt.tight_layout() #Prevents overlapping!     
```

## Histograms!

```python
# Histogram Subplot of Numeric Columns for Lap 1:
fig = plt.figure(figsize=(16,16))

#Make for loop that will plot Numeric Columns

for i, col in enumerate(lap1_num):
    fig.add_subplot(10,4, 1+i)
    sns.distplot(lap1_num[col], bins=45, kde=False, norm_hist=False, color='b')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('Count')    

plt.tight_layout() #Prevents overlapping!    
```

```python
# Histogram Subplot of Numeric Columns for Lap 2:
fig = plt.figure(figsize=(16,16))

#Make for loop that will plot Numeric Columns

for i, col in enumerate(lap2_num):
    fig.add_subplot(10,4, 1+i)
    sns.distplot(lap2_num[col], bins=100, kde=False, norm_hist=False, color='g')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('Count')    

plt.tight_layout() #Prevents overlapping!     
```

```python
# Histogram Subplot of Numeric Columns for Lap 3:
fig = plt.figure(figsize=(16,16))

#Make for loop that will plot Numeric Columns

for i, col in enumerate(lap3_num):
    fig.add_subplot(10,4, 1+i)
    sns.distplot(lap3_num[col], bins=60, kde=False, norm_hist=False, color='r')
    plt.title(col)
    plt.xlabel(col)
    plt.ylabel('Count')    

plt.tight_layout() #Prevents overlapping!  
```

Intersting observations so far:
* The third lap was the fastest speed, with best mean power (326 watts) and lowest mean cadence (88 RPM).
* On the third lap especially, the athlete seems to have some pedaling imbalances (ie: the Pedaling Delay Logs of Left vs. Right have greater differences).
* Mean Waist Angle Log Center increases by 2 degrees over the 3 laps.  Maybe the handlebar position is too low and/or the athlete has too much pressure on the front of the saddle?
* Mean Torso Angle also increases each lap.  This also supports revisiting the handlebar position and saddle choice.
* Mean Torso Rotation increases nearly 2 degrees.  How is this athlete's core strength?
